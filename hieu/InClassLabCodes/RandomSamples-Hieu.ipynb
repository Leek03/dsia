```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# create 11 random samples 
samples = np.random.randn(11)
print(samples)
mean = np.mean(samples[:10])
std = np.std(samples[:10])
print(mean,std)
```

```python
counter = 0
for i in range(10):
    samples[10] = np.random.randn(1)
    if abs(samples[10]-mean)>= 2*(std):
        plt.plot(samples)
        plt.show()
        counter += 1
        print('I,''ve found outlier')
print(counter)

        
```

# Step 3_1:test data to features and outcome

```python
X =input_data.drop(['diagnosis'], axis=1)
X.head().T
```

# Step 4: Define a model

```python
from sklearn.linear_model import LogisticRegression,Ridge,PassiveAggressiveRegressor,RidgeCV
from sklearn.datasets import make_regression,load_diabetes,load_iris
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
import numpy as np
```

```python
# My_model = LogisticRegression()
# new_model = LabelEncoder() # LabelEncoder buildIn classifications. 
y = LabelEncoder().fit_transform(input_data['diagnosis'])
# y.iloc[0:3,:]
# print(y.iloc[0:3,:])
# print(pd.DataFrame(y).iloc[0:19,:].T,'\n',pd.DataFrame(y).iloc[19:38,:].T)
pd.DataFrame(y).iloc[0:20,:].T #want to only print 20 instead of the couple of houndred if "y" and ".T" transpose.
```

# Step 5: Train the model & Other step Split data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.2,random_state=42) #,random_state=42
My_model = LogisticRegression()
My_model.fit(X_train,y_train)
```

# Step 6: Do the prediction on test data

```python
y_pred = My_model.predict(X_test)
pd.DataFrame(y_pred).iloc[0:20,:].T
```

# Step 7: Measure the accuracy

```python
from sklearn.metrics import accuracy_score
# accuracy_score(y_true, y_pred)
accuracy_score(y_test, y_pred)
```

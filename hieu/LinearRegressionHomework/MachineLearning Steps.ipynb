```python
import pandas as pd
```

```python
# Step 1_1: reading data
df_0=pd.read_csv('world-happiness-report/2017.csv')
df_0.shape
```

```python
# Step 1_2:Cleaning data
df = df_0.drop(['Country'], axis=1) # use inplace=True
```

```python
# Step 1_3:Saved cleaned data & Export to csv file:
df.to_csv('cleaned_2017.csv')
df=pd.read_csv('cleaned_2017.csv')
df.head()
```

```python
# Step 2: Features selection Drop columns( # remove unnecessary column/s)
df = df.drop(['Unnamed: 0'], axis =1)
print(df.columns,df.shape)
```

```python
#Look at the correlation heatmap to determine feature selection:
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.subplots(figsize = (11,11)) # (figsize = (18,18) size of the matrix
sns.set(font_scale=1.5)
sns.heatmap(df.corr(),square = True,cbar=True,annot=True,annot_kws={'size': 16})
#size correlaton eg1,-0.99
plt.show()
```

```python
#Change columns names 
sel_df = df[['Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.','Happiness.Rank']]
sel_df.columns = ['economy','family','health_expect','happy']
```

```python
# Step 3_1: Split data to train & test
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(sel_df, test_size= 0.20, random_state = 42)
df_train.head()
```

```python
# Step 3_2: Split train & test data to features & output/outcome
sel_df_train_feat = df_train.drop(['happy'], axis=1)
sel_df_train_output = df_train['happy']
print(sel_df_train_feat.head(3),'\n',sel_df_train_output.head(3))
```

```python
sel_df_test_feat =df_test.drop(['happy'], axis=1)
sel_df_test_output = df_test['happy']
print(sel_df_test_feat.head(3),'\n',sel_df_test_output.head().head(3))
```

```python
# Step 4: Define model
from sklearn.linear_model import LinearRegression,Ridge,PassiveAggressiveRegressor,RidgeCV
from sklearn.datasets import make_regression,load_diabetes
from sklearn import linear_model
import numpy as np
# My_model = LinearRegression()
# My_model = linear_model.Lasso(alpha=0.1)
My_model = Ridge(alpha=1.0)
# My_model = RidgeCV(alphas=[1e-3, 1e-2, 1e-1, 1]).fit(sel_df_train_feat,sel_df_train_output)
#for RidgeCV it step 4 & 5 are include as one.
```

```python
# Step 5: Train the model
My_model.fit(sel_df_train_feat,sel_df_train_output)  #SEE ABOVE
```

```python
# Step 6: Do the prediction on test data
y_pred = My_model.predict(sel_df_test_feat)
y_pred
```

```python
# Step 7: Measure the accuracy
from sklearn.metrics import mean_squared_error
print ('mean_squared_error= ', mean_squared_error(sel_df_test_output,y_pred))
from sklearn.metrics import r2_score
print ('r2_score= ', r2_score(sel_df_test_output,y_pred))
```
